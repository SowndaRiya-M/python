```python
#8th
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
data = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df = data.copy()
d= {}
def sep(iname): 
    l = [i.strip() for i in str(iname).split(',')]
    for j in l: 
        d[j] = d.get(j,0)+1 
    return 

df['InvestorsName'].apply(sep)  

m = max(list(d.values()))
for i in d: 
    if d[i] == m: 
        print(i, d[i])
```

```python
#6th
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df_start = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df_start['AmountInUSD'].fillna(0,inplace=True)
df_start['StartupName'].replace("Flipkart.com","Flipkart",inplace=True)
df_start['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
df_start['StartupName'].replace("Ola Cabs","Ola",inplace=True)
df_start['StartupName'].replace("Olacabs","Ola",inplace=True)
df_start['StartupName'].replace("Oyo Rooms","Oyo",inplace=True)
df_start['StartupName'].replace("Oyorooms","Oyo",inplace=True)
df_start['StartupName'].replace("OyoRooms","Oyo",inplace=True)
df_start['StartupName'].replace("OYO Rooms","Oyo",inplace=True)
df_start['AmountInUSD']=df_start['AmountInUSD'].apply(lambda x:float(str(x).replace(",","")))
df_start['AmountInUSD']=pd.to_numeric(df_start['AmountInUSD'])
start=df_start.groupby('StartupName')['AmountInUSD'].sum().sort_values(ascending=False)[0:5]
x=start.index
y=start.values
l=np.sum(y)
plt.pie(y,labels=x,autopct="%.2f%%")
plt.show()
for i in x:
    print(i)
```

```python
#5th
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df_start = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df_start['IndustryVertical'].replace("eCommerce","Ecommerce",inplace=True)
df_start['IndustryVertical'].replace("ECommerce","Ecommerce",inplace=True)
df_start['IndustryVertical'].replace("ecommerce","Ecommerce",inplace=True)
df_start['AmountInUSD']=df_start['AmountInUSD'].apply(lambda x:float(str(x).replace(",","")))
df_start['AmountInUSD']=pd.to_numeric(df_start['AmountInUSD'])
industry=df_start.groupby('IndustryVertical')['AmountInUSD'].sum().sort_values(ascending=False)[0:5]
x=industry.index
y=industry.values
l=np.sum(y)
plt.pie(y,labels=x,autopct="%.2f%%")
plt.show()
for i,j in zip(x,y):
    print(i,format((j/l)*100, '.2f'))
```

```python
#3rd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df_start = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df_start['CityLocation'].dropna(inplace=True)
df_start['AmountInUSD'].dropna(inplace=True)
def separateCity(city):
    return city.split('/')[0].strip()
df_start['CityLocation']=df_start['CityLocation'].apply(separateCity)
df_start['CityLocation'].replace("Delhi","New Delhi",inplace=True)
df_start['CityLocation'].replace("bangalore","Bangalore",inplace=True)
df_start['AmountInUSD']=df_start['AmountInUSD'].apply(lambda x:float(str(x).replace(",","")))
df_start['AmountInUSD']=pd.to_numeric(df_start['AmountInUSD'])
city_number=df_start.groupby('CityLocation')['AmountInUSD'].sum().sort_values(ascending=False)[0:10]
city=city_number.index
fundvalue=city_number.values
y=sum(fundvalue)
ind=np.where(fundvalue)
newcity=city[ind]
for i,j in zip(newcity,fundvalue):
    print(i,format((j/y)*100, '.2f'))
```

```python
#2nd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df_start = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df_start['CityLocation'].dropna(inplace=True)
def separateCity(city):
    return city.split('/')[0].strip()
df_start['CityLocation']=df_start['CityLocation'].apply(separateCity)
df_start['CityLocation'].replace("Delhi","New Delhi",inplace=True)
df_start['CityLocation'].replace("bangalore","Bangalore",inplace=True)
city_number=df_start['CityLocation'].value_counts()[0:10]
city=city_number.index
numCity=city_number.values
plt.pie(numCity,labels=city,autopct="%.2f%%")
for i in range(len(city)):
    print(city[i],numCity[i])
```

```python
#1st one
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df = pd.read_csv('C:/Users/sowndariya/Desktop/CNN/startup_funding.csv')
df['Date'].replace("12/05.2015","12/05/2015",inplace=True) 
df['Date'].replace("13/04.2015","13/04/2015",inplace=True) 
df['Date'].replace("15/01.2015","15/01/2015",inplace=True) 
df['Date'].replace("22/01//2015","22/01/2015",inplace=True) 
year = df['Date'].str.split('/',expand=True)[2]
freq = year.value_counts(ascending=True)
freq.sort_index(inplace=True)
x = freq.index
y = freq.values
plt.plot(x,y)
plt.title('year vs fundings')
plt.xlabel('year')
plt.ylabel('no of fundings')
plt.show()
for i in range(len(x)):
    print(x[i],y[i])
```
