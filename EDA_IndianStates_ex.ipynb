```python
print(data.sort_values('sex_ratio',ascending=False)['name_of_city'].head(1))

```

### City with lowest Sex- Ratio

```python
print(data.sort_values('sex_ratio',ascending=True)['name_of_city'].head(1))


```

### Mean of Sex- Ratio grouped by States and sorted by sex ratio in ascending order.

```python
print(data.columns)
data.groupby('state_name')[['sex_ratio']].mean().sort_values('sex_ratio',ascending=True)
```

### Shows top 5 states having lowest sex-ratio. It should sorted by state name.

```python
state=data.groupby('state_name')[['sex_ratio']].min().sort_values(by='sex_ratio',ascending=True).head(5)
state.sort_values(by='state_name',ascending=True)
```

```python
data.groupby('state_name')[['sex_ratio']].max().sort_values(by=['sex_ratio','state_name'],ascending=[True,True]).head(5)
```

```python
data[['state_name','sex_ratio']].sort_values(by=['sex_ratio','state_name'],ascending=[True,True]).head(5).sort_values(by='state_name')
```

### 5 Cities, which have lowest sex- ratio and making a dataframe out of it.

```python
mini=data.groupby('name_of_city')[['sex_ratio']].min().sort_values(by='sex_ratio',ascending=True).head(5)
mini
```

### 5 Cities, which have highest sex- ratio and making a dataframe out of it.

```python
maxi=data.groupby('name_of_city')[['sex_ratio']].max().sort_values(by='sex_ratio',ascending=False).head(5)
maxi
```

### Concatenated top 5 and bottom 5 cities on the basis of Sex - Ratio

```python
pd.concat([mini,maxi])
```
