```python
import pandas as pd
crime=pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
crime
```

```python
#Convert the type of the column Year to datetime64
crime['Year'] = pd.to_datetime(crime.Year, format='%Y')
crime.info()
crime.head()
```

```python
#Set the Year column as the index of the dataframe¶
crime = crime.set_index('Year', drop = True)
crime.head()
```

```python
crime.head()
```

```python
#Group the year by decades and sum the values¶
s=crime.groupby('Year').sum()
s
```

```python
#Group the year by decades and sum the values¶
a=crime.resample('10AS').sum()
a
```

```python
# Uses resample to get the max value only for the "Population" column
pop=a['Population'].resample('10AS').max()
pop
```

```python
# Updating the "Population" column
a['Population']=pop
a
```

```python
users =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/users.csv')
users
sessions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/sessions.csv')
products =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/products.csv')
transactions =pd.read_csv('https://raw.githubusercontent.com/ben519/DataWrangling/master/Data/transactions.csv')
print(users.columns)
print(transactions.columns)
```

```python
#Join users to transactions, keeping all rows from transactions and only matching rows fromusers (left join)
pd.merge(transactions,users,how='left',left_on='UserID',right_on='UserID')
print(users)
transactions
```

```python
#Which transactions have a UserID not in users?
a=pd.merge(transactions,users)
```

```python
#Join users to transactions, keeping only rows from transactions and users that match via UserID (inner join)
pd.merge(transactions,users,how='inner',on='UserID')
```

```python
#Join users to transactions, displaying all matching rows AND all non-matching rows (full outer join)
pd.merge(transactions,users,how='outer',on='UserID')
```
