# Q8) Is there any ingredient which has remained in the top 10 most common ingredients across all years?

```python
import pandas as pd
dff=pd.read_csv("Hair Care Data 14.09.2019.csv")
dff.astype('object')
dff.dtypes
dff.columns
dff['year']=dff['Date Published'].apply(lambda x: str(x).split('-')[2])
dff['year']=dff['year'].astype('int')

listt1,listt2,listt3,listt4,listt5,listt6,listt7,listt8,listt9=[],[],[],[],[],[],[],[],[]
dff.dropna(subset = ['Ingredient 1', 'Ingredient 2', 'Ingredient 3',
       'Ingredient 4', 'Ingredient 5', 'Ingredient 6', 'Ingredient 7',
       'Ingredient 8', 'Ingredient 9','Remaining Ingredients'],inplace = True)
for row in dff['Ingredient 1']:
    listt1.append(row)
for row in dff['Ingredient 2']:
    listt2.append(row)
for row in dff['Ingredient 3']:
    listt3.append(row)
for row in dff['Ingredient 4']:
    listt4.append(row)
for row in dff['Ingredient 5']:
    listt5.append(row)
for row in dff['Ingredient 6']:
    listt6.append(row)
for row in dff['Ingredient 7']:
    listt7.append(row)  
for row in dff['Ingredient 8']:
    listt8.append(row)   
for row in dff['Ingredient 9']:
    listt9.append(row) 


new=[]
for i in listt9:
    if (i in listt1 and i in listt3 and i in listt4 and i in listt5 and i in listt6 and i in listt7 and i in listt8 and i in listt9):
        new.append(i)

aa=dff['year'].value_counts()
yr=list(aa.index)
dictt={}
for yr in dff['year']:
    for j in new:
        if j in dictt:
            dictt[j]+=1
        else:
            dictt[j]=1
print("TOP 10 MOST COMMON INGREDIENTS ACROSS ALL YEARS")
keys=[]        
for key, value in sorted(dictt.items(), key=lambda item: item[1],reverse=True)[0:10]:
    keys.append(key)
    print(key,end=",")
new_rem=[]
for k in dff['Remaining Ingredients']:
    if k in keys:
        new_rem.append(k)
print('\n')
print('REMAINED INGRDIENTS IN TOP 10 INGREDIENTS ACROSS ALL YEARS')
print(list(set(new_rem)))
    
```
